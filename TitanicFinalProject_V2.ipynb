 ''' AGEWISE SURVIVAL OF FEMALES AND MALES '''

```python

titanic = titanic [[ "SurvivedCategory", "Sex" , "Age" ]]

females = titanic.loc[titanic["Sex"] == "female"]
survived_f = females.loc[females["SurvivedCategory"]=="Survived"]
f_age = survived_f[["Age"]]
clean_f_age = f_age.dropna(how = "any")
clean_f_age.head()
```

```python
clean_f_age = np.array(clean_f_age)
```

```python

group1 =[]
group2 =[]
group3 =[]

for x in clean_f_age:
    if (x <= 25):
        group1.append(x)
    elif (x <= 50):
        group2.append(x)
    else:
        group3.append(x)
         
  
```

```python
group1 = len(group1)
group2 = len(group2)
group3 = len(group3)
```

```python

xaxis =  ["age(0 to 25)" , "age(26 to 50)" , "age(51 and above)"]
tick_locations = [x for x in xaxis]


age_groups = [group1 ,group2 , group3]


plt.figure(figsize= (20,10))

plt.bar( xaxis, age_groups , alpha = 0.75, align = "center", color=[ 'orchid', 'magenta', 'mediumvioletred'] )

plt.title("FEMALE SURVIVAL ON THE TITANIC BY AGE" , size = 30)
plt.xlabel("Age of Female Survivors" , size = 25 , color = "red")
plt.ylabel("Number of Female Survivors" , size= 25 , color = "red")

plt.xlim(-0.75,len(xaxis)-0.25)
plt.ylim(0, max(age_groups) + 10)

plt.xticks( tick_locations, xaxis , rotation= 45 , size = 18 , color = "crimson")
plt.yticks(size = 18 , color = "crimson")
plt.grid(alpha= 0.25)
plt.show()



```

''' CLASSWISE SURVIVAL RATE OF FEMALES AND MALES '''

```python

titanic = titanic [["SurvivedCategory", "Class" ,"Sex"]]
females = titanic.loc[titanic["Sex"] == "female"]
survived_f = females.loc[females["SurvivedCategory"]=="Survived"]
class_f = survived_f["Class"].value_counts()

class_f
```

```python

class_nos = [ 91,72,70 ]
labels = ["First Class" ,"Coach" , "Business Class" ]
colors= [ "blue" , "mediumorchid", "blueviolet"]
explode = (0.10,  0 , 0)

plt.figure(figsize=(30,20))
plt.title("% OF TOTAL FEMALES SURVIVED BY CLASS TYPE" , size =35, color ="m")
plt.pie(class_nos, explode=explode, labels=labels ,colors=colors, autopct="%1.1f%%", shadow=True, startangle=180,
        textprops={'fontsize': 30})
plt.axis("equal")
plt.show

```
